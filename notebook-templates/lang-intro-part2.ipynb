## 1.  Statements & Control Structures



### Assignment



```python
# simple, single target assignment

a = 0
b = 'hello'
```

```python
# can also assign to target "lists"

a, b, c = 0, 'hello', True
```

```python
# note: expression on right is fully evaluated, then are assigned to
#       elements in the "target" list, from left to right

x, y, z = 1, 2, 3
x, y, z = x+y, y+z, x+y+z
```

```python
# easy python "swap"

a, b = 'apples', 'bananas'
a, b = b, a
```

```python
# note: order matters!

a, b, a = 1, 2, 3
print(a, b)
```

```python
# can also have multiple assignments in a row -- consistent with
# above: expression is evaluated first, then assigned to all targets
# from left to right (note: order matters!)

x = y = z = None
print(x, y, z)
```

```python
it = iter(r)
while True:
    try:
        x = next(it)
        print(x)
    except StopIteration:
        break
```

```python
import math

def quadratic_roots(a, b, c):
    disc = b**2-4*a*c
    if disc < 0:
        return None
    else:
        return (-b+math.sqrt(disc))/(2*a), (-b-math.sqrt(disc))/(2*a)
```

```python
quadratic_roots(1, -5, 6) # eq = (x-3)(x-2)
```

```python
quadratic_roots(a=1, b=-5, c=6)
```

```python
quadratic_roots(c=6, a=1, b=-5)
```

```python
class Shape:
    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return self.name

    def __str__(self):
        return self.name.upper()

    def area(self):
        raise NotImplementedError()
```

```python
s = Shape('circle')
print(s.__str__())
```

```python
s
```

```python
str(s)
```

```python
try:
    s.area()
except NotImplementedError as e:
    print(e)
```

```python
class Circle(Shape):
    def __init__(self, radius):
        super().__init__('circle')
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2
```

```python
c = Circle(5.0)
c
c.area()
```

```python
class Circle(Shape):
    def __init__(self, radius):
        super().__init__('circle')
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

    def __eq__(self, other):
        return isinstance(other, Circle) and self.radius == other.radius

    def __add__(self, other):
        return Circle(self.radius + other.radius)

    def __repr__(self):
        return 'Circle(r={})'.format(self.radius)
```

```python
c1 = Circle(2.0)
c2 = Circle(4.0)
c3 = Circle(2.0)

c1, c2, c3
c1 == c2
c1 == c3
c1 + c2
```
