## 1.  The `LinkedList` and `Node` classes



```python
class LinkedList:
    class Node:
        def __init__(self, val, next=None):
            self.val = val
            self.next = next

    def __init__(self):
        self.count = 0
        self.head = None ##SENTINEL ("dummy") head
        self.head.next = self.head
        self.head.prior = self.head

    def prepend(self, value):
        self.count += 1
        newn = self.Node(value, prior = self.head, next = self.head.next)  ##prior pointer is previous head, next pointer is node after the head
        self.head.next.prior = newn
        self.head.next = newn

    def append(self, value): #O(1)
        self.count += 1
        newn = self.Node(value, prior = self.head.prior, next = self.head)
        ##seems confusing but is actually simple. Easiser to think of these examples with DLLs (doubly linked lists) that are relatively long (3+ elements.) The prior pointer of an element appended to the END of a DLL is...the prior element of the element at the FRONT of the list (which is just self.head.prior. The next pointer of an element appended to the end of a list is...the element at the front of the list (self.head))
        self.head.prior.next = newn #again...seems simple but easy when you break it down. What's the prior of the head of the ORIGINAL list (before "newn" came along?) Yep - it's at the end of the list. That ending "next" at the end of the list is then set to point to "newn". 
        self.head.prior = newn

    def __getitem__(self, idx): #n = O(n), but we can do it in n/2
        #Write n/2 (if it is in the first half of the list, we access through  "next"; if it is in the second half we access through "prior")
        assert(idx >= 0 and idx < len(self))
        n = self.head.next
        for i in range(0, idx):
            n = n.next
        return n.val

    def __len__(self):
        return self.count

    def __iter__(self):
        n = self.head
        while n:
            yield n.val
            n = n.next

    def __repr__(self):
        return '[' + ', '.join(x for x in self) + ']'
```

```python
lst = LinkedList()
for i in range(10):
    lst.prepend(i)
lst
```

## 1.  Bidirectional links (Doubly-linked list) & Sentinel head



```python
class LinkedList:
    class Node:
        def __init__(self, val, prior=None, next=None):
            self.val = val
            self.prior = prior
            self.next  = next

    def __init__(self):
        self.count = 0

    def prepend(self, value):
        self.count += 1

    def append(self, value):
        self.count += 1

    def __getitem__(self, idx):
        pass

    def __len__(self):
        return self.count

    def __iter__(self):
        n = self.head.next
        while n is not self.head:
            yield n.val
            n = n.next

    def __repr__(self):
        return '[' + ', '.join(str(x) for x in self) + ']'
```

```python
lst = LinkedList()
for i in range(10):
    lst.prepend(i)
for i in range(10):
    lst.append(i)
lst
```

## 1.  Run-time analysis



Run-time complexities for circular, doubly-linked list of $N$ elements:

-   indexing (position-based access) = $O(?)$
-   search (unsorted) = $O(?)$
-   search (sorted) = $O(?)$ &#x2014; binary search isn't possible!
-   prepend = $O(?)$
-   append = $O(?)$
-   indexing = $O(?)$

-   deletion of arbitrary element: indexing = $O(?)$ + deletion = $O(?)$
